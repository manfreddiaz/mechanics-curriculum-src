```python
import os
import json
import pandas as pd
import seaborn as sns
from IPython.display import display
```

```python
RESULTS = {
    "sum" : {
        "<center>MinAtar</center>[DQN] [Shapl]": "propt/sum/minatar/random/dqn/shapley",
        "<center>MinAtar</center>[DQN] [S & B]": "propt/sum/minatar/ordered/dqn/sanchez_bergantinos",
        "<center>MinAtar</center>[DQN] [N & R]": "propt/sum/minatar/ordered/dqn/nowak_radzik",
        "<center>sIPD</center>[PPO] [Shapl]": "propt/sum/sipd/random/ppo/shapley",
        "<center>sIPD</center>[PPO] [S & B]": "propt/sum/sipd/ordered/ppo/sanchez_bergantinos",
        "<center>sIPD</center>[PPO] [N & R]": "propt/sum/sipd/ordered/ppo/nowak_radzik",
    },
    "mean": {
        "<center>MinAtar</center>[DQN] [Shapl]": "propt/mean/minatar/random/dqn/shapley",
        "<center>MinAtar</center>[DQN] [S & B]": "propt/mean/minatar/ordered/dqn/sanchez_bergantinos",
        "<center>MinAtar</center>[DQN] [N & R]": "propt/mean/minatar/ordered/dqn/nowak_radzik",
        "<center>sIPD</center>[PPO] [Shapl]": "propt/mean/sipd/random/ppo/shapley",
        "<center>sIPD</center>[PPO] [S & B]": "propt/mean/sipd/ordered/ppo/sanchez_bergantinos",
        "<center>sIPD</center>[PPO] [N & R]": "propt/mean/sipd/ordered/ppo/nowak_radzik",
    }
}
```

```python
def load_results(agg: str):
    dfs = [
        pd.read_json(os.path.join(folder, 'game.json')) for _, folder in RESULTS[agg].items()]
    return pd.concat(dfs, keys=RESULTS[agg].keys())
```

```python
def render_propt(agg: str):
    df = load_results(agg)
    df = df[['players', 'additive', 'probs']]
    props = 'font-size:2em;'
    for idx, r in df.groupby(level=0):
        display(r.style.background_gradient(cmap='viridis').applymap(lambda x: props).applymap_index(lambda x: props))
    
    return df
```
